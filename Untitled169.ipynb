```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

def voltage_waveform(t):
    V1 = 23.2
    V2 = 23.4
    V3 = 23.6
    V4 = 23.4
    V5 = 23.2

    T1, T2, T3, T4 = 0, 1, 2, 3
    t_mod = t % 4

    if t_mod < T2:
        return V2
    elif t_mod < T3:
        return V3
    elif t_mod < T4:
        return V4
    else:
        return V5

# Generate data for given G and T with Iout
def generate_data(G, T, noise_level=0.0, N=None):
    import numpy as np
    import torch

    Rsh_true = 93.0571
    Rs_true = 0.2185
    Ceq_true = 0.2496
    Iph_true = 8.09

    # Time span
    t = [0.0, 0.002, 0.004, 0.006, 0.008, 0.01, 0.012, 0.014, 0.016, 0.018, 0.02, 0.022, 0.024, 0.026, 0.028, 0.03, 0.032, 0.034, 0.036, 0.038, 0.04, 0.042, 0.044, 0.046, 0.048, 0.05, 0.052, 0.054, 0.056, 0.058, 0.06, 0.062, 0.064, 0.066, 0.068, 0.07, 0.072, 0.074, 0.076, 0.078, 0.08, 0.082, 0.084, 0.086, 0.088, 0.09, 0.092, 0.094, 0.096, 0.098, 0.1, 0.102, 0.105, 0.107, 0.109, 0.111, 0.113, 0.115, 0.117, 0.119, 0.121, 0.123, 0.125, 0.127, 0.129, 0.131, 0.133, 0.135, 0.137, 0.139, 0.141, 0.143, 0.145, 0.147, 0.149, 0.151, 0.153, 0.155, 0.157, 0.159, 0.161, 0.163, 0.165, 0.167, 0.169, 0.171, 0.173, 0.175, 0.177, 0.179, 0.181, 0.183, 0.185, 0.187, 0.189, 0.191, 0.193, 0.195, 0.197, 0.199, 0.201, 0.203, 0.205, 0.208, 0.21, 0.212, 0.214, 0.216, 0.218, 0.22, 0.222, 0.224, 0.226, 0.228, 0.23, 0.232, 0.234, 0.236, 0.238, 0.24, 0.242, 0.244, 0.246, 0.248, 0.25, 0.252, 0.254, 0.256, 0.258, 0.26, 0.262, 0.264, 0.266, 0.268, 0.27, 0.272, 0.274, 0.276, 0.278, 0.28, 0.282, 0.284, 0.286, 0.288, 0.29, 0.292, 0.294, 0.296, 0.298, 0.3, 0.302, 0.304, 0.306, 0.308, 0.311, 0.313, 0.315, 0.317, 0.319, 0.321, 0.323, 0.325, 0.327, 0.329, 0.331, 0.333, 0.335, 0.337, 0.339, 0.341, 0.343, 0.345, 0.347, 0.349, 0.351, 0.353, 0.355, 0.357, 0.359, 0.361, 0.363, 0.365, 0.367, 0.369, 0.371, 0.373, 0.375, 0.377, 0.379, 0.381, 0.383, 0.385, 0.387, 0.389, 0.391, 0.393, 0.395, 0.397, 0.399, 0.401, 0.403, 0.405, 0.407, 0.409, 0.411, 0.413, 0.416, 0.418, 0.42, 0.422, 0.424, 0.426, 0.428, 0.43, 0.432, 0.434, 0.436, 0.438, 0.44, 0.442, 0.444, 0.446, 0.448, 0.45, 0.452, 0.454, 0.456, 0.458, 0.46, 0.462, 0.464, 0.466, 0.468, 0.47, 0.472, 0.474, 0.476, 0.478, 0.48, 0.482, 0.484, 0.486, 0.488, 0.49, 0.492, 0.494, 0.496, 0.498, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5000001, 0.5000007, 0.5000018, 0.5000127, 0.50007254, 0.5001813, 0.50126934, 0.50281334, 0.5048134, 0.50681335, 0.5088134, 0.51081336, 0.5128134, 0.51481336, 0.51681334, 0.5188134, 0.52081335, 0.5228134, 0.52481335, 0.5268134, 0.52881336, 0.53081334, 0.5328134, 0.53481334, 0.5368134, 0.53881335, 0.5408134, 0.54281336, 0.54481333, 0.54681337, 0.54881334, 0.5508134, 0.55281335, 0.5548134, 0.55681336, 0.5588134, 0.56081337, 0.56281334, 0.5648134, 0.56681335, 0.5688134, 0.57081336, 0.5728134, 0.57481337, 0.57681334, 0.5788134, 0.58081335, 0.5828134, 0.58481336, 0.5868134, 0.58981335, 0.5918134, 0.59381336, 0.5958134, 0.59781337, 0.59981334, 0.6018134, 0.60381335, 0.6058134, 0.60781336, 0.6098134, 0.61181337, 0.61381334, 0.6158134, 0.61781335, 0.6198134, 0.62181336, 0.6238134, 0.62581336, 0.62781334, 0.6298134, 0.63181335, 0.6338134, 0.63581336, 0.6378134, 0.63981336, 0.64181334, 0.6438134, 0.64581335, 0.6478134, 0.64981335, 0.6518134, 0.65381336, 0.65581334, 0.6578134, 0.65981334, 0.6618134, 0.66381335, 0.6658134, 0.66781336, 0.66981333, 0.67181337, 0.67381334, 0.6758134, 0.67781335, 0.6798134, 0.68181336, 0.6838134, 0.68581337, 0.68781334, 0.6898134, 0.69281334, 0.6948134, 0.69681334, 0.6988134, 0.70081335, 0.7028134, 0.70481336, 0.70681334, 0.70881337, 0.71081334, 0.7128134, 0.71481335, 0.7168134, 0.71881336, 0.7208134, 0.72281337, 0.72481334, 0.7268134, 0.72881335, 0.7308134, 0.73281336, 0.7348134, 0.73681337, 0.73881334, 0.7408134, 0.74281335, 0.7448134, 0.74681336, 0.7488134, 0.75081336, 0.75281334, 0.7548134, 0.75681335, 0.7588134, 0.76081336, 0.7628134, 0.76481336, 0.76681334, 0.7688134, 0.77081335, 0.7728134, 0.77481335, 0.7768134, 0.77881336, 0.78081334, 0.7828134, 0.78481334, 0.7868134, 0.78881335, 0.7908134, 0.79281336, 0.79481333, 0.79781336, 0.7998134, 0.80181336, 0.80381334, 0.8058134, 0.80781335, 0.8098134, 0.81181335, 0.8138134, 0.81581336, 0.81781334, 0.8198134, 0.82181334, 0.8238134, 0.82581335, 0.8278134, 0.82981336, 0.83181334, 0.83381337, 0.83581334, 0.8378134, 0.83981335, 0.8418134, 0.84381336, 0.8458134, 0.84781337, 0.84981334, 0.8518134, 0.85381335, 0.8558134, 0.85781336, 0.8598134, 0.86181337, 0.86381334, 0.8658134, 0.86781335, 0.8698134, 0.87181336, 0.8738134, 0.87581336, 0.87781334, 0.8798134, 0.88181335, 0.8838134, 0.88581336, 0.8878134, 0.88981336, 0.89181334, 0.8938134, 0.89581335, 0.8978134, 0.90081334, 0.9028134, 0.90481335, 0.9068134, 0.90881336, 0.9108134, 0.91281337, 0.91481334, 0.9168134, 0.91881335, 0.9208134, 0.92281336, 0.9248134, 0.92681336, 0.92881334, 0.9308134, 0.93281335, 0.9348134, 0.93681335, 0.9388134, 0.94081336, 0.94281334, 0.9448134, 0.94681334, 0.9488134, 0.95081335, 0.9528134, 0.95481336, 0.95681334, 0.95881337, 0.96081334, 0.9628134, 0.96481335, 0.9668134, 0.96881336, 0.9708134, 0.97281337, 0.97481334, 0.9768134, 0.97881335, 0.9808134, 0.98281336, 0.9848134, 0.98681337, 0.98881334, 0.9908134, 0.99281335, 0.9948134, 0.99681336, 0.9988134, 1.0008134, 1.0038134, 1.0058134, 1.0078133, 1.0098133, 1.0118134, 1.0138134, 1.0158134, 1.0178133, 1.0198134, 1.0218134, 1.0238134, 1.0258133, 1.0278133, 1.0298134, 1.0318134, 1.0338134, 1.0358133, 1.0378133, 1.0398134, 1.0418134, 1.0438133, 1.0458133, 1.0478134, 1.0498134, 1.0518134, 1.0538133, 1.0558133, 1.0578134, 1.0598134, 1.0618134, 1.0638133, 1.0658134, 1.0678134, 1.0698134, 1.0718133, 1.0738133, 1.0758134, 1.0778134, 1.0798134, 1.0818133, 1.0838133, 1.0858134, 1.0878134, 1.0898134, 1.0918133, 1.0938134, 1.0958134, 1.0978134, 1.0998133, 1.1018133, 1.1038134, 1.1068133, 1.1088134, 1.1108134, 1.1128134, 1.1148133, 1.1168134, 1.1188134, 1.1208134, 1.1228133, 1.1248133, 1.1268134, 1.1288134, 1.1308134, 1.1328133, 1.1348133, 1.1368134, 1.1388134, 1.1408134, 1.1428133, 1.1448134, 1.1468134, 1.1488134, 1.1508133, 1.1528133, 1.1548134, 1.1568134, 1.1588134, 1.1608133, 1.1628133, 1.1648134, 1.1668134, 1.1688133, 1.1708133, 1.1728134, 1.1748134, 1.1768134, 1.1788133, 1.1808133, 1.1828134, 1.1848134, 1.1868134, 1.1888133, 1.1908134, 1.1928134, 1.1948134, 1.1968133, 1.1988133, 1.2008134, 1.2028134, 1.2048134, 1.2068133, 1.2088133, 1.2118133, 1.2138133, 1.2158134, 1.2178134, 1.2198133, 1.2218133, 1.2238134, 1.2258134, 1.2278134, 1.2298133, 1.2318133, 1.2338134, 1.2358134, 1.2378134, 1.2398133, 1.2418134, 1.2438134, 1.2458134, 1.2478133, 1.2498133, 1.2518134, 1.2538134, 1.2558134, 1.2578133, 1.2598133, 1.2618134, 1.2638134, 1.2658134, 1.2678133, 1.2698134, 1.2718134, 1.2738134, 1.2758133, 1.2778133, 1.2798134, 1.2818134, 1.2838134, 1.2858133, 1.2878133, 1.2898134, 1.2918134, 1.2938133, 1.2958133, 1.2978134, 1.2998134, 1.3018134, 1.3038133, 1.3058133, 1.3078134, 1.3098134, 1.3118134, 1.3148134, 1.3168133, 1.3188133, 1.3208134, 1.3228134, 1.3248134, 1.3268133, 1.3288133, 1.3308134, 1.3328134, 1.3348134, 1.3368133, 1.3388133, 1.3408134, 1.3428134, 1.3448133, 1.3468133, 1.3488134, 1.3508134, 1.3528134, 1.3548133, 1.3568133, 1.3588134, 1.3608134, 1.3628134, 1.3648133, 1.3668134, 1.3688134, 1.3708134, 1.3728133, 1.3748133, 1.3768134, 1.3788134, 1.3808134, 1.3828133, 1.3848133, 1.3868134, 1.3888134, 1.3908134, 1.3928133, 1.3948134, 1.3968134, 1.3988134, 1.4008133, 1.4028133, 1.4048134, 1.4068134, 1.4088134, 1.4108133, 1.4128133, 1.4148134, 1.4178134, 1.4198134, 1.4218134, 1.4238133, 1.4258133, 1.4278134, 1.4298134, 1.4318134, 1.4338133, 1.4358133, 1.4378134, 1.4398134, 1.4418133, 1.4438133, 1.4458134, 1.4478134, 1.4498134, 1.4518133, 1.4538133, 1.4558134, 1.4578134, 1.4598134, 1.4618133, 1.4638133, 1.4658134, 1.4678134, 1.4698133, 1.4718133, 1.4738134, 1.4758134, 1.4778134, 1.4798133, 1.4818133, 1.4838134, 1.4858134, 1.4878134, 1.4898133, 1.4918134, 1.4938134, 1.4958134, 1.4978133, 1.4998133, 1.5018134, 1.5038134, 1.5058134, 1.5078133, 1.5098133, 1.5118134, 1.5138134, 1.5158134, 1.5178133, 1.5208133, 1.5228133, 1.5248134, 1.5268134, 1.5288134, 1.5308133, 1.5328133, 1.5348134, 1.5368134, 1.5388134, 1.5408133, 1.5428134, 1.5448134, 1.5468134, 1.5488133, 1.5508133, 1.5528134, 1.5548134, 1.5568134, 1.5588133, 1.5608133, 1.5628134, 1.5648134, 1.5668133, 1.5688133, 1.5708134, 1.5728134, 1.5748134, 1.5768133, 1.5788133, 1.5808134, 1.5828134, 1.5848134, 1.5868133, 1.5888133, 1.5908134, 1.5928134, 1.5948133, 1.5968133, 1.5988134, 1.6008134, 1.6028134, 1.6048133, 1.6068133, 1.6088134, 1.6108134, 1.6128134, 1.6148133, 1.6168134, 1.6188134, 1.6208134, 1.6228133, 1.6258134, 1.6278133, 1.6298133, 1.6318134, 1.6338134, 1.6358134, 1.6378133, 1.6398134, 1.6418134, 1.6438134, 1.6458133, 1.6478133, 1.6498134, 1.6518134, 1.6538134, 1.6558133, 1.6578133, 1.6598134, 1.6618134, 1.6638134, 1.6658133, 1.6678134, 1.6698134, 1.6718134, 1.6738133, 1.6758133, 1.6778134, 1.6798134, 1.6818134, 1.6838133, 1.6858133, 1.6878134, 1.6898134, 1.6918133, 1.6938133, 1.6958134, 1.6978134, 1.6998134, 1.7018133, 1.7038133, 1.7058134, 1.7078134, 1.7098134, 1.7118133, 1.7138133, 1.7158134, 1.7178134, 1.7198133, 1.7218133, 1.7238134, 1.7258134, 1.7288134, 1.7308134, 1.7328134, 1.7348133, 1.7368133, 1.7388134, 1.7408134, 1.7428133, 1.7448133, 1.7468134, 1.7488134, 1.7508134, 1.7528133, 1.7548133, 1.7568134, 1.7588134, 1.7608134, 1.7628133, 1.7648134, 1.7668134, 1.7688134, 1.7708133, 1.7728133, 1.7748134, 1.7768134, 1.7788134, 1.7808133, 1.7828133, 1.7848134, 1.7868134, 1.7888134, 1.7908133, 1.7928134, 1.7948134, 1.7968134, 1.7988133, 1.8008133, 1.8028134, 1.8048134, 1.8068134, 1.8088133, 1.8108133, 1.8128134, 1.8148134, 1.8168133, 1.8188133, 1.8208134, 1.8228134, 1.8248134, 1.8268133, 1.8288133, 1.8318133, 1.8338133, 1.8358134, 1.8378134, 1.8398134, 1.8418133, 1.8438134, 1.8458134, 1.8478134, 1.8498133, 1.8518133, 1.8538134, 1.8558134, 1.8578134, 1.8598133, 1.8618133, 1.8638134, 1.8658134, 1.8678133, 1.8698133, 1.8718134, 1.8738134, 1.8758134, 1.8778133, 1.8798133, 1.8818134, 1.8838134, 1.8858134, 1.8878133, 1.8898134, 1.8918134, 1.8938134, 1.8958133, 1.8978133, 1.8998134, 1.9018134, 1.9038134, 1.9058133, 1.9078133, 1.9098134, 1.9118134, 1.9138134, 1.9158133, 1.9178134, 1.9198134, 1.9218134, 1.9238133, 1.9258133, 1.9278134, 1.9298134, 1.9318134, 1.9348134, 1.9368134, 1.9388133, 1.9408134, 1.9428134, 1.9448134, 1.9468133, 1.9488133, 1.9508134, 1.9528134, 1.9548134, 1.9568133, 1.9588133, 1.9608134, 1.9628134, 1.9648134, 1.9668133, 1.9688134, 1.9708134, 1.9728134, 1.9748133, 1.9768133, 1.9788134, 1.9808134, 1.9828134, 1.9848133, 1.9868133, 1.9888134, 1.9908134, 1.9928133, 1.9948133, 1.9968134, 1.9988134, 2.0, 2.0003734, 2.00156, 2.0035598, 2.00556, 2.00756, 2.0095599, 2.01156, 2.0135598, 2.01556, 2.01756, 2.0195599, 2.02156, 2.0235598, 2.02556, 2.02756, 2.0295599, 2.03156, 2.0335598, 2.0365598, 2.03856, 2.04056, 2.0425599, 2.04456, 2.0465598, 2.04856, 2.05056, 2.0525599, 2.05456, 2.0565598, 2.05856, 2.06056, 2.0625598, 2.06456, 2.0665598, 2.06856, 2.07056, 2.0725598, 2.07456, 2.07656, 2.0785599, 2.08056, 2.0825598, 2.08456, 2.08656, 2.0885599, 2.09056, 2.0925598, 2.09456, 2.09656, 2.0985599, 2.10056, 2.1025598, 2.10456, 2.10656, 2.1085598, 2.11056, 2.1125598, 2.11456, 2.11656, 2.1185598, 2.12056, 2.12256, 2.1245599, 2.12656, 2.1285598, 2.13056, 2.13256, 2.1345599, 2.13656, 2.13956, 2.1415598, 2.14356, 2.14556, 2.14756, 2.14956, 2.1515598, 2.15356, 2.15556, 2.1575599, 2.15956, 2.1615598, 2.16356, 2.16556, 2.1675599, 2.16956, 2.1715598, 2.17356, 2.17556, 2.1775599, 2.17956, 2.1815598, 2.18356, 2.18556, 2.1875598, 2.18956, 2.1915598, 2.19356, 2.19556, 2.1975598, 2.19956, 2.20156, 2.2035599, 2.20556, 2.2075598, 2.20956, 2.21156, 2.2135599, 2.21556, 2.2175598, 2.21956, 2.22156, 2.2235599, 2.22556, 2.2275598, 2.22956, 2.23156, 2.2335598, 2.23556, 2.2375598, 2.23956, 2.24256, 2.24456, 2.2465599, 2.24856, 2.2505598, 2.25256, 2.25456, 2.2565598, 2.25856, 2.2605598, 2.26256, 2.26456, 2.2665598, 2.26856, 2.27056, 2.27256, 2.27456, 2.2765598, 2.27856, 2.28056, 2.2825599, 2.28456, 2.2865598, 2.28856, 2.29056, 2.2925599, 2.29456, 2.2965598, 2.29856, 2.30056, 2.3025599, 2.30456, 2.3065598, 2.30856, 2.31056, 2.3125598, 2.31456, 2.3165598, 2.31856, 2.32056, 2.3225598, 2.32456, 2.32656, 2.3285599, 2.33056, 2.3325598, 2.33456, 2.33656, 2.3385599, 2.34056, 2.3425598, 2.3455598, 2.34756, 2.34956, 2.3515599, 2.35356, 2.3555598, 2.35756, 2.35956, 2.3615599, 2.36356, 2.3655598, 2.36756, 2.36956, 2.3715599, 2.37356, 2.3755598, 2.37756, 2.37956, 2.3815598, 2.38356, 2.3855598, 2.38756, 2.38956, 2.3915598, 2.39356, 2.39556, 2.39756, 2.39956, 2.4015598, 2.40356, 2.40556, 2.4075599, 2.40956, 2.4115598, 2.41356, 2.41556, 2.4175599, 2.41956, 2.4215598, 2.42356, 2.42556, 2.4275599, 2.42956, 2.4315598, 2.43356, 2.43556, 2.4375598, 2.43956, 2.4415598, 2.44356, 2.44556, 2.4475598, 2.4505599, 2.45256, 2.4545598, 2.45656, 2.45856, 2.4605598, 2.46256, 2.4645598, 2.46656, 2.46856, 2.4705598, 2.47256, 2.47456, 2.4765599, 2.47856, 2.4805598, 2.48256, 2.48456, 2.4865599, 2.48856, 2.4905598, 2.49256, 2.49456, 2.4965599, 2.49856, 2.5005598, 2.50256, 2.50456, 2.5065598, 2.50856, 2.5105598, 2.51256, 2.51456, 2.5165598, 2.51856, 2.52056, 2.52256, 2.52456, 2.5265598, 2.52856, 2.53056, 2.5325599, 2.53456, 2.5365598, 2.53856, 2.54056, 2.5425599, 2.54456, 2.5465598, 2.54856, 2.55056, 2.55356, 2.5555599, 2.55756, 2.5595598, 2.56156, 2.56356, 2.5655599, 2.56756, 2.5695598, 2.57156, 2.57356, 2.5755599, 2.57756, 2.5795598, 2.58156, 2.58356, 2.5855598, 2.58756, 2.5895598, 2.59156, 2.59356, 2.5955598, 2.59756, 2.59956, 2.6015599, 2.60356, 2.6055598, 2.60756, 2.60956, 2.6115599, 2.61356, 2.6155598, 2.61756, 2.61956, 2.6215599, 2.62356, 2.6255598, 2.62756, 2.62956, 2.6315598, 2.63356, 2.6355598, 2.63756, 2.63956, 2.6415598, 2.64356, 2.64556, 2.64756, 2.64956, 2.6515598, 2.65356, 2.65656, 2.6585598, 2.66056, 2.66256, 2.6645598, 2.66656, 2.6685598, 2.67056, 2.67256, 2.6745598, 2.67656, 2.67856, 2.6805599, 2.68256, 2.6845598, 2.68656, 2.68856, 2.6905599, 2.69256, 2.6945598, 2.69656, 2.69856, 2.7005599, 2.70256, 2.7045598, 2.70656, 2.70856, 2.7105598, 2.71256, 2.7145598, 2.71656, 2.71856, 2.7205598, 2.72256, 2.72456, 2.7265599, 2.72856, 2.7305598, 2.73256, 2.73456, 2.7365599, 2.73856, 2.7405598, 2.74256, 2.74456, 2.7465599, 2.74856, 2.7505598, 2.75256, 2.75456, 2.7565598, 2.7595599, 2.76156, 2.7635598, 2.76556, 2.76756, 2.7695599, 2.77156, 2.7735598, 2.77556, 2.77756, 2.7795599, 2.78156, 2.7835598, 2.78556, 2.78756, 2.7895598, 2.79156, 2.7935598, 2.79556, 2.79756, 2.7995598, 2.80156, 2.80356, 2.8055599, 2.80756, 2.8095598, 2.81156, 2.81356, 2.8155599, 2.81756, 2.8195598, 2.82156, 2.82356, 2.8255599, 2.82756, 2.8295598, 2.83156, 2.83356, 2.8355598, 2.83756, 2.8395598, 2.84156, 2.84356, 2.8455598, 2.84756, 2.84956, 2.8515599, 2.85356, 2.8555598, 2.85756, 2.85956, 2.8615599, 2.86456, 2.86656, 2.8685598, 2.87056, 2.87256, 2.8745599, 2.87656, 2.8785598, 2.88056, 2.88256, 2.8845599, 2.88656, 2.8885598, 2.89056, 2.89256, 2.8945599, 2.89656, 2.8985598, 2.90056, 2.90256, 2.9045599, 2.90656, 2.9085598, 2.91056, 2.91256, 2.9145598, 2.91656, 2.9185598, 2.92056, 2.92256, 2.9245598, 2.92656, 2.92856, 2.9305599, 2.93256, 2.9345598, 2.93656, 2.93856, 2.9405599, 2.94256, 2.9445598, 2.94656, 2.94856, 2.9505599, 2.95256, 2.9545598, 2.95656, 2.95856, 2.9605598, 2.96256, 2.9645598, 2.9675598, 2.96956, 2.97156, 2.9735599, 2.97556, 2.9775598, 2.97956, 2.98156, 2.9835598, 2.98556, 2.9875598, 2.98956, 2.99156, 2.9935598, 2.99556, 2.99756, 2.9995599, 3.00156, 3.0035598, 3.00556, 3.00756, 3.0095599, 3.01156, 3.0135598, 3.01556, 3.01756, 3.0195599, 3.02156, 3.0235598, 3.02556, 3.02756, 3.0295599, 3.03156, 3.0335598, 3.03556, 3.03756, 3.0395598, 3.04156, 3.0435598, 3.04556, 3.04756, 3.0495598, 3.05156, 3.05356, 3.0555599, 3.05756, 3.0595598, 3.06156, 3.06356, 3.0655599, 3.06756, 3.07056, 3.0725598, 3.07456, 3.07656, 3.0785599, 3.08056, 3.0825598, 3.08456, 3.08656, 3.0885599, 3.09056, 3.0925598, 3.09456, 3.09656, 3.0985599, 3.10056, 3.1025598, 3.10456, 3.10656, 3.1085598, 3.11056, 3.1125598, 3.11456, 3.11656, 3.1185598, 3.12056, 3.12256, 3.1245599, 3.12656, 3.1285598, 3.13056, 3.13256, 3.1345599, 3.13656, 3.1385598, 3.14056, 3.14256, 3.1445599, 3.14656, 3.1485598, 3.15056, 3.15256, 3.1545599, 3.15656, 3.1585598, 3.16056, 3.16256, 3.1645598, 3.16656, 3.1685598, 3.17056, 3.17356, 3.17556, 3.1775599, 3.17956, 3.1815598, 3.18356, 3.18556, 3.1875598, 3.18956, 3.1915598, 3.19356, 3.19556, 3.1975598, 3.19956, 3.20156, 3.2035599, 3.20556, 3.2075598, 3.20956, 3.21156, 3.2135599, 3.21556, 3.2175598, 3.21956, 3.22156, 3.2235599, 3.22556, 3.2275598, 3.22956, 3.23156, 3.2335598, 3.23556, 3.2375598, 3.23956, 3.24156, 3.2435598, 3.24556, 3.24756, 3.2495599, 3.25156, 3.2535598, 3.25556, 3.25756, 3.2595599, 3.26156, 3.2635598, 3.26556, 3.26756, 3.2695599, 3.27156, 3.2735598, 3.27556, 3.27856, 3.28056, 3.2825599, 3.28456, 3.2865598, 3.28856, 3.29056, 3.2925599, 3.29456, 3.2965598, 3.29856, 3.30056, 3.3025599, 3.30456, 3.3065598, 3.30856, 3.31056, 3.3125598, 3.31456, 3.3165598, 3.31856, 3.32056, 3.3225598, 3.32456, 3.32656, 3.3285599, 3.33056, 3.3325598, 3.33456, 3.33656, 3.3385599, 3.34056, 3.3425598, 3.34456, 3.34656, 3.3485599, 3.35056, 3.3525598, 3.35456, 3.35656, 3.3585598, 3.36056, 3.3625598, 3.36456, 3.36656, 3.3685598, 3.37056, 3.37256, 3.3745599, 3.37656, 3.3785598, 3.3815598, 3.38356, 3.3855598, 3.38756, 3.38956, 3.3915598, 3.39356, 3.39556, 3.39756, 3.39956, 3.4015598, 3.40356, 3.40556, 3.4075599, 3.40956, 3.4115598, 3.41356, 3.41556, 3.4175599, 3.41956, 3.4215598, 3.42356, 3.42556, 3.4275599, 3.42956, 3.4315598, 3.43356, 3.43556, 3.4375598, 3.43956, 3.4415598, 3.44356, 3.44556, 3.4475598, 3.44956, 3.45156, 3.4535599, 3.45556, 3.4575598, 3.45956, 3.46156, 3.4635599, 3.46556, 3.4675598, 3.46956, 3.47156, 3.4735599, 3.47556, 3.4775598, 3.47956, 3.48156, 3.48456, 3.4865599, 3.48856, 3.4905598, 3.49256, 3.49456, 3.4965599, 3.49856, 3.5, 3.5008802, 3.5023203, 3.5043204, 3.5063202, 3.5083203, 3.5103202, 3.5123203, 3.5143204, 3.5163202, 3.5183203, 3.5203202, 3.5223203, 3.5243204, 3.5263202, 3.5283203, 3.5303202, 3.5323203, 3.5343204, 3.5363202, 3.5383203, 3.5403204, 3.5423203, 3.5443203, 3.5463202, 3.5483203, 3.5503204, 3.5523202, 3.5543203, 3.5563202, 3.5583203, 3.5603204, 3.5623202, 3.5643203, 3.5663202, 3.5683203, 3.5703204, 3.5723202, 3.5743203, 3.5763202, 3.5783203, 3.5803204, 3.5823202, 3.5853202, 3.5873203, 3.5893202, 3.5913203, 3.5933204, 3.5953202, 3.5973203, 3.5993202, 3.6013203, 3.6033204, 3.6053202, 3.6073203, 3.6093202, 3.6113203, 3.6133204, 3.6153202, 3.6173203, 3.6193204, 3.6213202, 3.6233203, 3.6253202, 3.6273203, 3.6293204, 3.6313202, 3.6333203, 3.6353202, 3.6373203, 3.6393204, 3.6413202, 3.6433203, 3.6453202, 3.6473203, 3.6493204, 3.6513202, 3.6533203, 3.6553202, 3.6573203, 3.6593204, 3.6613202, 3.6633203, 3.6653204, 3.6673203, 3.6693203, 3.6713202, 3.6733203, 3.6753204, 3.6773202, 3.6793203, 3.6813202, 3.6833203, 3.6853204, 3.6873202, 3.6903203, 3.6923203, 3.6943202, 3.6963203, 3.6983204, 3.7003202, 3.7023203, 3.7043202, 3.7063203, 3.7083204, 3.7103202, 3.7123203, 3.7143202, 3.7163203, 3.7183204, 3.7203202, 3.7223203, 3.7243202, 3.7263203, 3.7283204, 3.7303202, 3.7323203, 3.7343202, 3.7363203, 3.7383204, 3.7403202, 3.7423203, 3.7443204, 3.7463202, 3.7483203, 3.7503202, 3.7523203, 3.7543204, 3.7563202, 3.7583203, 3.7603202, 3.7623203, 3.7643204, 3.7663202, 3.7683203, 3.7703202, 3.7723203, 3.7743204, 3.7763202, 3.7783203, 3.7803202, 3.7823203, 3.7843204, 3.7863202, 3.7883203, 3.7903204, 3.7933202, 3.7953203, 3.7973204, 3.7993202, 3.8013203, 3.8033202, 3.8053203, 3.8073204, 3.8093202, 3.8113203, 3.8133202, 3.8153203, 3.8173203, 3.8193202, 3.8213203, 3.8233204, 3.8253202, 3.8273203, 3.8293202, 3.8313203, 3.8333204, 3.8353202, 3.8373203, 3.8393202, 3.8413203, 3.8433204, 3.8453202, 3.8473203, 3.8493202, 3.8513203, 3.8533204, 3.8553202, 3.8573203, 3.8593202, 3.8613203, 3.8633204, 3.8653202, 3.8673203, 3.8693204, 3.8713202, 3.8733203, 3.8753202, 3.8773203, 3.8793204, 3.8813202, 3.8833203, 3.8853202, 3.8873203, 3.8893204, 3.8913202, 3.8933203, 3.8963203, 3.8983202, 3.9003203, 3.9023204, 3.9043202, 3.9063203, 3.9083202, 3.9103203, 3.9123204, 3.9143202, 3.9163203, 3.9183202, 3.9203203, 3.9223204, 3.9243202, 3.9263203, 3.9283202, 3.9303203, 3.9323204, 3.9343202, 3.9363203, 3.9383202, 3.9403203, 3.9423203, 3.9443202, 3.9463203, 3.9483204, 3.9503202, 3.9523203, 3.9543202, 3.9563203, 3.9583204, 3.9603202, 3.9623203, 3.9643202, 3.9663203, 3.9683204, 3.9703202, 3.9723203, 3.9743202, 3.9763203, 3.9783204, 3.9803202, 3.9823203, 3.9843202, 3.9863203, 3.9883204, 3.9903202, 3.9923203, 3.9943204, 3.9963202, 3.9993203]

    # Input voltage (Vin)
    Vin = np.array([voltage_waveform(ti) for ti in t])

    # Original Iout without noise
    Iout = [6.920939, 6.9527493, 6.9838686, 7.0138817, 7.0428185, 7.0707164, 7.0976133, 7.1235447, 7.1485457, 7.1726494, 7.195888, 7.2182927, 7.239893, 7.2607183, 7.2807965, 7.3001537, 7.3188167, 7.3368096, 7.354157, 7.3708816, 7.387006, 7.4025517, 7.4175396, 7.4319897, 7.445921, 7.4593525, 7.472302, 7.4847865, 7.4968233, 7.5084276, 7.519616, 7.5304027, 7.540802, 7.5508285, 7.560495, 7.5698147, 7.5787997, 7.5874624, 7.595814, 7.603866, 7.611629, 7.619114, 7.6263294, 7.6332865, 7.6399937, 7.64646, 7.6526947, 7.658705, 7.6645, 7.6700873, 7.675474, 7.6806674, 7.68811, 7.692849, 7.697419, 7.7018247, 7.7060723, 7.7101674, 7.7141156, 7.7179217, 7.7215915, 7.7251296, 7.728541, 7.7318296, 7.7350006, 7.738057, 7.7410045, 7.743846, 7.7465854, 7.7492266, 7.751773, 7.754228, 7.7565947, 7.758877, 7.761077, 7.763198, 7.765243, 7.7672143, 7.7691154, 7.770948, 7.7727146, 7.774418, 7.7760606, 7.7776437, 7.77917, 7.780642, 7.782061, 7.783429, 7.784748, 7.7860193, 7.7872453, 7.7884274, 7.789567, 7.7906656, 7.7917247, 7.792746, 7.7937303, 7.7946796, 7.7955947, 7.796477, 7.7973275, 7.7981477, 7.7989383, 7.8000717, 7.800793, 7.801489, 7.80216, 7.8028064, 7.80343, 7.804031, 7.8046107, 7.805169, 7.805708, 7.806227, 7.806728, 7.807211, 7.8076763, 7.808125, 7.8085575, 7.8089747, 7.8093767, 7.8097644, 7.810138, 7.8104987, 7.810846, 7.811181, 7.811504, 7.8118153, 7.8121157, 7.812405, 7.812684, 7.812953, 7.8132124, 7.8134623, 7.8137035, 7.8139358, 7.81416, 7.814376, 7.8145843, 7.814785, 7.8149786, 7.815165, 7.8153453, 7.815519, 7.8156857, 7.8158474, 7.816003, 7.8161526, 7.816297, 7.8164363, 7.8165708, 7.8167005, 7.8168254, 7.8169456, 7.817118, 7.817228, 7.8173337, 7.817436, 7.8175344, 7.8176293, 7.817721, 7.817809, 7.817894, 7.817976, 7.818055, 7.8181314, 7.818205, 7.818276, 7.818344, 7.81841, 7.818474, 7.818535, 7.818594, 7.8186507, 7.8187056, 7.8187585, 7.8188095, 7.8188586, 7.8189063, 7.8189516, 7.818996, 7.8190384, 7.8190794, 7.819119, 7.8191566, 7.8191934, 7.819229, 7.819263, 7.819296, 7.819328, 7.8193583, 7.819388, 7.819416, 7.8194437, 7.81947, 7.819495, 7.81952, 7.819544, 7.8195662, 7.8195887, 7.8196096, 7.81963, 7.8196497, 7.819669, 7.8196874, 7.819705, 7.8197303, 7.8197465, 7.8197618, 7.819777, 7.8197913, 7.819805, 7.8198185, 7.8198314, 7.8198442, 7.819856, 7.8198676, 7.819879, 7.8198895, 7.8199, 7.81991, 7.8199196, 7.819929, 7.819938, 7.819947, 7.8199553, 7.819963, 7.819971, 7.819978, 7.819986, 7.8199925, 7.819999, 7.820006, 7.820012, 7.820018, 7.820024, 7.8200293, 7.820035, 7.82004, 7.820045, 7.82005, 7.8200545, 7.820059, 7.820063, 7.8200674, 7.8200717, 7.8200755, 7.8200793, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.8200827, 7.820083, 7.820085, 7.8200874, 7.820091, 7.820094, 7.820097, 7.8201003, 7.820103, 7.8201056, 7.8201084, 7.820111, 7.820113, 7.8201156, 7.820118, 7.8201203, 7.8201222, 7.8201246, 7.8201265, 7.8201284, 7.8201303, 7.820132, 7.8201337, 7.820135, 7.820137, 7.8201385, 7.82014, 7.8201413, 7.8201427, 7.820144, 7.820145, 7.8201466, 7.8201475, 7.820149, 7.82015, 7.820151, 7.820152, 7.820153, 7.820154, 7.8201547, 7.8201556, 7.8201566, 7.8201575, 7.820158, 7.820159, 7.82016, 7.820161, 7.8201613, 7.8201623, 7.820163, 7.8201632, 7.8201637, 7.8201647, 7.820165, 7.8201656, 7.820166, 7.8201666, 7.820167, 7.8201675, 7.820168, 7.8201685, 7.8201685, 7.820169, 7.8201694, 7.82017, 7.8201704, 7.8201704, 7.820171, 7.8201714, 7.8201714, 7.820172, 7.820172, 7.8201723, 7.820173, 7.820173, 7.8201733, 7.8201733, 7.8201737, 7.8201737, 7.8201737, 7.820174, 7.820174, 7.8201747, 7.8201747, 7.8201747, 7.820175, 7.820175, 7.820175, 7.8201756, 7.8201756, 7.8201756, 7.820176, 7.820176, 7.820176, 7.820176, 7.8201766, 7.8201766, 7.8201766, 7.8201766, 7.820177, 7.820177, 7.820177, 7.820177, 7.820177, 7.8201776, 7.8201776, 7.8201776, 7.8201776, 7.8201776, 7.8201776, 7.8201776, 7.820178, 7.820178, 7.820178, 7.820178, 7.820178, 7.820178, 7.820178, 7.820178, 7.820178, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.8201785, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.820179, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 6.9187946, 6.966303, 6.996868, 7.026338, 7.0547504, 7.0821433, 7.1085534, 7.134015, 7.1585636, 7.1822305, 7.2050486, 7.2270474, 7.2482567, 7.268705, 7.2884192, 7.3074265, 7.3257513, 7.343418, 7.360451, 7.376873, 7.392706, 7.40797, 7.4226866, 7.436875, 7.450554, 7.4637423, 7.476457, 7.4887156, 7.500534, 7.5119286, 7.5229144, 7.533506, 7.543717, 7.5535617, 7.563053, 7.572204, 7.5810266, 7.5895324, 7.5977325, 7.605639, 7.6132617, 7.62061, 7.6276956, 7.6345267, 7.6411123, 7.647462, 7.6535835, 7.6594853, 7.6651754, 7.670661, 7.67595, 7.6810493, 7.688357, 7.693011, 7.697498, 7.7018237, 7.705994, 7.7100153, 7.713892, 7.7176294, 7.721233, 7.7247066, 7.728056, 7.7312856, 7.734399, 7.7374, 7.740294, 7.743084, 7.745774, 7.7483673, 7.7508674, 7.7532783, 7.755602, 7.7578425, 7.760003, 7.7620854, 7.7640934, 7.7660294, 7.7678957, 7.7696953, 7.77143, 7.7731028, 7.774715, 7.77627, 7.7777686, 7.779214, 7.780607, 7.7819505, 7.783245, 7.784494, 7.7856975, 7.786858, 7.787977, 7.789056, 7.790096, 7.7910986, 7.792065, 7.7929974, 7.7938957, 7.794762, 7.7955976, 7.7964025, 7.7971787, 7.7979274, 7.7990003, 7.799683, 7.800342, 7.8009768, 7.801589, 7.8021793, 7.802748, 7.803297, 7.803826, 7.804336, 7.8048277, 7.8053017, 7.8057585, 7.806199, 7.806624, 7.8070335, 7.8074284, 7.807809, 7.808176, 7.80853, 7.808871, 7.8092, 7.809517, 7.8098226, 7.8101172, 7.8104014, 7.8106756, 7.81094, 7.8111944, 7.81144, 7.8116765, 7.811905, 7.8121247, 7.812337, 7.8125415, 7.8127384, 7.8129287, 7.813112, 7.8132887, 7.813459, 7.813623, 7.8137813, 7.8139343, 7.814081, 7.8142233, 7.81436, 7.8144917, 7.814619, 7.8147416, 7.81486, 7.814974, 7.8151374, 7.8152413, 7.8153415, 7.8154383, 7.8155313, 7.8156214, 7.8157077, 7.8157916, 7.8158717, 7.8159494, 7.8160243, 7.816097, 7.8161664, 7.816233, 7.816298, 7.8163605, 7.8164206, 7.8164783, 7.816534, 7.816588, 7.81664, 7.81669, 7.8167386, 7.816785, 7.8168297, 7.816873, 7.816915, 7.816955, 7.8169937, 7.8170314, 7.817067, 7.817102, 7.8171353, 7.8171678, 7.8171988, 7.817229, 7.817258, 7.8172855, 7.8173127, 7.8173385, 7.8173637, 7.8173876, 7.817411, 7.8174334, 7.817455, 7.817476, 7.817496, 7.8175154, 7.817534, 7.817552, 7.8175693, 7.817594, 7.81761, 7.817625, 7.81764, 7.817654, 7.817668, 7.817681, 7.8176937, 7.817706, 7.817718, 7.817729, 7.81774, 7.817751, 7.817761, 7.817771, 7.8177805, 7.8177896, 7.817798, 7.8178067, 7.817815, 7.817823, 7.8178306, 7.8178377, 7.817845, 7.817852, 7.817858, 7.817865, 7.817871, 7.817877, 7.8178825, 7.817888, 7.817893, 7.8178983, 7.817903, 7.817908, 7.8179126, 7.817917, 7.817921, 7.8179255, 7.8179293, 7.817933, 7.817937, 7.81794, 7.8179436, 7.817947, 7.8179502, 7.817953, 7.817956, 7.817959, 7.8179617, 7.8179646, 7.8179684, 7.8179708, 7.817973, 7.817975, 7.8179774, 7.8179793, 7.8179812, 7.817983, 7.817985, 7.817987, 7.817989, 7.8179903, 7.817992, 7.8179936, 7.817995, 7.8179965, 7.817998, 7.8179994, 7.818001, 7.8180017, 7.818003, 7.818004, 7.818005, 7.8180065, 7.8180075, 7.8180084, 7.8180094, 7.8180103, 7.8180113, 7.818012, 7.8180127, 7.8180137, 7.8180146, 7.818015, 7.818016, 7.8180165, 7.8180175, 7.818018, 7.8180184, 7.8180194, 7.81802, 7.8180203, 7.818021, 7.8180213, 7.818022, 7.8180223, 7.8180227, 7.818023, 7.8180237, 7.818024, 7.8180246, 7.818025, 7.8180256, 7.818026, 7.818026, 7.8180265, 7.818027, 7.818027, 7.8180275, 7.8180275, 7.818028, 7.8180285, 7.8180285, 7.818029, 7.818029, 7.8180294, 7.8180294, 7.81803, 7.81803, 7.81803, 7.8180304, 7.8180304, 7.818031, 7.818031, 7.818031, 7.8180313, 7.8180313, 7.8180313, 7.8180313, 7.818032, 7.818032, 7.818032, 7.8180323, 7.8180323, 7.8180323, 7.8180323, 7.8180327, 7.8180327, 7.8180327, 7.8180327, 7.8180327, 7.818033, 7.818033, 7.818033, 7.818033, 7.818033, 7.818033, 7.818033, 7.8180337, 7.8180337, 7.8180337, 7.8180337, 7.8180337, 7.8180337, 7.8180337, 7.818034, 7.818034, 7.818034, 7.818034, 7.818034, 7.818034, 7.818034, 7.818034, 7.818034, 7.818034, 7.818034, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.8180346, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 7.818035, 8.730731, 8.724513, 8.705107, 8.67335, 8.642732, 8.613212, 8.584751, 8.557312, 8.530858, 8.505353, 8.480763, 8.457056, 8.4342, 8.412164, 8.390919, 8.370436, 8.350688, 8.33165, 8.313293, 8.286988, 8.270235, 8.254084, 8.238511, 8.223498, 8.209024, 8.195069, 8.181615, 8.168644, 8.156138, 8.144081, 8.132457, 8.12125, 8.110445, 8.100028, 8.089985, 8.080302, 8.070967, 8.061966, 8.053289, 8.044923, 8.036858, 8.029081, 8.0215845, 8.014357, 8.007388, 8.0006695, 7.9941916, 7.987947, 7.981926, 7.976121, 7.970525, 7.965129, 7.959927, 7.9549117, 7.9500766, 7.945415, 7.9409204, 7.9365873, 7.93241, 7.928382, 7.9244986, 7.920755, 7.9171453, 7.9136653, 7.9103103, 7.907076, 7.9039574, 7.9009504, 7.8980517, 7.895257, 7.891252, 7.8887014, 7.8862424, 7.8838716, 7.8815856, 7.879382, 7.8772573, 7.875209, 7.873234, 7.87133, 7.8694944, 7.8677244, 7.8660183, 7.864373, 7.8627872, 7.861258, 7.8597836, 7.858362, 7.8569922, 7.855671, 7.8543973, 7.8531694, 7.8519855, 7.850844, 7.8497434, 7.8486824, 7.8476596, 7.8466735, 7.8457227, 7.8448057, 7.843922, 7.84307, 7.8422484, 7.8414564, 7.840693, 7.8399568, 7.8392467, 7.8385625, 7.837903, 7.837267, 7.8366537, 7.8360624, 7.835492, 7.834943, 7.834413, 7.8339024, 7.83341, 7.832935, 7.832477, 7.8320355, 7.83161, 7.8310003, 7.830612, 7.830238, 7.829877, 7.829529, 7.829193, 7.82887, 7.828558, 7.828257, 7.827967, 7.8276877, 7.8274183, 7.8271585, 7.826908, 7.826667, 7.8264337, 7.8262095, 7.825993, 7.825784, 7.825583, 7.8253894, 7.8252025, 7.825022, 7.824848, 7.824681, 7.824519, 7.824363, 7.8242135, 7.8240685, 7.823929, 7.8237944, 7.8236647, 7.8235397, 7.823419, 7.8233027, 7.8231907, 7.8230824, 7.8229785, 7.822878, 7.822781, 7.8226876, 7.8225975, 7.8225107, 7.8224273, 7.8223467, 7.822269, 7.8221936, 7.8221216, 7.822052, 7.821985, 7.82192, 7.821827, 7.821768, 7.821711, 7.8216558, 7.821603, 7.821552, 7.8215027, 7.821455, 7.821409, 7.8213654, 7.821323, 7.8212814, 7.8212423, 7.821204, 7.821167, 7.8211317, 7.8210974, 7.8210645, 7.821033, 7.8210025, 7.820973, 7.8209443, 7.8209167, 7.8208904, 7.8208647, 7.8208404, 7.8208165, 7.8207936, 7.8207717, 7.82075, 7.8207297, 7.82071, 7.820691, 7.820673, 7.820655, 7.820638, 7.8206215, 7.8206058, 7.8205905, 7.8205757, 7.8205614, 7.8205476, 7.8205347, 7.820522, 7.8205094, 7.8204975, 7.820486, 7.820475, 7.8204646, 7.8204546, 7.8204446, 7.820435, 7.820421, 7.8204126, 7.8204045, 7.8203964, 7.8203883, 7.820381, 7.8203735, 7.820367, 7.82036, 7.8203535, 7.8203473, 7.820341, 7.8203354, 7.82033, 7.8203244, 7.820319, 7.8203144, 7.8203096, 7.820305, 7.8203, 7.820296, 7.820292, 7.8202877, 7.820284, 7.82028, 7.8202767, 7.820273, 7.8202696, 7.8202662, 7.8202634, 7.8202605, 7.8202577, 7.820255, 7.820252, 7.8202496, 7.8202467, 7.8202443, 7.820242, 7.82024, 7.8202376, 7.8202357, 7.820234, 7.8202314, 7.82023, 7.820228, 7.820226, 7.820225, 7.820223, 7.8202214, 7.82022, 7.8202186, 7.820216, 7.820215, 7.820214, 7.8202124, 7.8202114, 7.82021, 7.820209, 7.820208, 7.820207, 7.820206, 7.820205, 7.8202043, 7.8202033, 7.8202024, 7.8202014, 7.820201, 7.8202, 7.8201995, 7.8201985, 7.820198, 7.820197, 7.8201966, 7.820196, 7.820195, 7.8201947, 7.8201942, 7.820194, 7.8201933, 7.820193, 7.8201923, 7.820192, 7.8201914, 7.820191, 7.8201904, 7.82019, 7.82019, 7.8201895, 7.820189, 7.8201885, 7.8201885, 7.820188, 7.8201876, 7.8201876, 7.820187, 7.820187, 7.8201866, 7.8201866, 7.820186, 7.820186, 7.8201857, 7.8201857, 7.820185, 7.820185, 7.8201847, 7.8201847, 7.820184, 7.820184, 7.820184, 7.8201838, 7.8201838, 7.8201838, 7.8201833, 7.8201833, 7.8201833, 7.820183, 7.820183, 7.820183, 7.820183, 7.8201823, 7.8201823, 7.8201823, 7.8201823, 7.8201823, 7.820182, 7.820182, 7.820182, 7.820182, 7.820182, 7.820182, 7.8201814, 7.8201814, 7.8201814, 7.8201814, 7.8201814, 7.8201814, 7.8201814, 7.820181, 7.820181, 7.820181, 7.820181, 7.820181, 7.820181, 7.820181, 7.820181, 7.820181, 7.820181, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.8201804, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.82018, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 7.8201795, 8.705687, 8.674056, 8.643491, 8.614021, 8.5856085, 8.558215, 8.531806, 8.506344, 8.481795, 8.458129, 8.43531, 8.413312, 8.392102, 8.3716545, 8.351939, 8.332932, 8.314608, 8.296941, 8.279907, 8.263486, 8.247653, 8.232389, 8.217672, 8.203485, 8.189805, 8.176617, 8.163902, 8.151644, 8.139825, 8.12843, 8.117445, 8.1068535, 8.0966425, 8.086798, 8.072689, 8.063704, 8.055041, 8.046689, 8.038637, 8.030874, 8.02339, 8.016173, 8.009216, 8.002509, 7.996043, 7.9898086, 7.983798, 7.978003, 7.972416, 7.967029, 7.961836, 7.956829, 7.952002, 7.947348, 7.942861, 7.938535, 7.934365, 7.930344, 7.9264674, 7.92273, 7.9191265, 7.9156523, 7.912303, 7.909074, 7.9059606, 7.902959, 7.900065, 7.897275, 7.894585, 7.8919916, 7.8894916, 7.887081, 7.884757, 7.8825164, 7.8803563, 7.8782735, 7.8762655, 7.8743296, 7.872463, 7.870664, 7.868929, 7.8672566, 7.865644, 7.8640895, 7.8625903, 7.860442, 7.859074, 7.857755, 7.856484, 7.855258, 7.854076, 7.8529363, 7.8518376, 7.8507786, 7.849757, 7.848773, 7.8478236, 7.8469086, 7.8460264, 7.8451757, 7.8443556, 7.843565, 7.8428025, 7.8420677, 7.841359, 7.840676, 7.8400173, 7.839382, 7.83877, 7.8381796, 7.8376107, 7.8370624, 7.836533, 7.8360233, 7.8355317, 7.8350577, 7.8346004, 7.83416, 7.833735, 7.833326, 7.832931, 7.83255, 7.8321834, 7.8318295, 7.831488, 7.831159, 7.830842, 7.8305364, 7.8302417, 7.8299575, 7.829684, 7.8294196, 7.829165, 7.8289194, 7.8286824, 7.8284545, 7.828234, 7.827919, 7.8277183, 7.8275247, 7.827338, 7.827158, 7.8269844, 7.8268175, 7.826656, 7.8265004, 7.8263507, 7.826206, 7.826067, 7.8259325, 7.825803, 7.825678, 7.8255577, 7.825442, 7.82533, 7.825222, 7.825118, 7.8250175, 7.824921, 7.8248277, 7.824738, 7.8246512, 7.824568, 7.824487, 7.8244095, 7.8243346, 7.8242626, 7.824193, 7.824126, 7.824061, 7.823999, 7.823939, 7.8238807, 7.823825, 7.823771, 7.823719, 7.823669, 7.823621, 7.823574, 7.8235292, 7.823486, 7.8234444, 7.823404, 7.823365, 7.823328, 7.823292, 7.823257, 7.8232236, 7.8231754, 7.823145, 7.8231153, 7.823087, 7.8230596, 7.8230333, 7.823008, 7.8229833, 7.8229594, 7.822937, 7.8229146, 7.8228936, 7.822873, 7.8228536, 7.8228345, 7.822816, 7.8227983, 7.8227816, 7.822765, 7.822749, 7.822734, 7.822719, 7.822705, 7.8226914, 7.822678, 7.822665, 7.8226533, 7.8226414, 7.82263, 7.822619, 7.8226085, 7.822598, 7.8225884, 7.822579, 7.8225694, 7.822561, 7.822552, 7.822544, 7.822536, 7.8225284, 7.822521, 7.822514, 7.8225074, 7.8225007, 7.8224945, 7.8224883, 7.822482, 7.8224764, 7.822471, 7.822466, 7.8224607, 7.8224535, 7.8224487, 7.8224444, 7.82244, 7.822436, 7.8224316, 7.8224277, 7.822424, 7.8224216, 7.82242, 7.8224173, 7.822414, 7.822411, 7.8224077, 7.822405, 7.822402, 7.822399, 7.8223963, 7.822394, 7.822391, 7.8223886, 7.8223863, 7.8223844, 7.822382, 7.82238, 7.8223777, 7.822376, 7.822374, 7.8223724, 7.8223705, 7.822369, 7.822367, 7.8223658, 7.8223643, 7.822363, 7.8223615, 7.82236, 7.8223586, 7.8223577, 7.822356, 7.8223553, 7.822354, 7.822353, 7.822352, 7.822351, 7.82235, 7.822349, 7.822348, 7.822347, 7.822346, 7.8223453, 7.8223443, 7.8223433, 7.822343, 7.8223424, 7.8223414, 7.822341, 7.82234, 7.8223395, 7.822339, 7.8223386, 7.822338, 7.8223376, 7.822337, 7.8223367, 7.822336, 7.8223357, 7.8223352, 7.822335, 7.8223343, 7.822334, 7.822334, 7.8223333, 7.822333, 7.8223324, 7.8223324, 7.822332, 7.822332, 7.8223314, 7.822331, 7.822331, 7.8223305, 7.8223305, 7.82233, 7.82233, 7.8223295, 7.8223295, 7.822329, 7.822329, 7.822329, 7.8223286, 7.8223286, 7.822328, 7.822328, 7.822328, 7.8223276, 7.8223276, 7.8223276, 7.8223276, 7.822327, 7.822327, 7.822327, 7.822327, 7.8223267, 7.8223267, 7.8223267, 7.8223267, 7.822326, 7.822326, 7.822326, 7.822326, 7.822326, 7.8223257, 7.8223257, 7.8223257, 7.8223257, 7.8223257, 7.8223257, 7.8223257, 7.822325, 7.822325, 7.822325, 7.822325, 7.822325, 7.822325, 7.822325, 7.822325, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223248, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.8223243, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324, 7.822324]
    # Convert to tensors
    t_combined = torch.tensor(t, dtype=torch.float32).view(-1, 1).requires_grad_(True)
    Iout_noisy = torch.tensor(Iout, dtype=torch.float32).view(-1, 1)
    Vin = torch.tensor(Vin, dtype=torch.float32).view(-1, 1)

    # Add noise to Iout_noisy if noise_level > 0
    if noise_level > 0:
        Iout_noisy += noise_level * torch.randn_like(Iout_noisy)

    # Sampling at N/2 rate
    if N is not None:
        sampling_indices = np.arange(0, len(t), len(t) // (N // 2))
        t_combined = t_combined[sampling_indices]
        Iout_noisy = Iout_noisy[sampling_indices]
        Vin = Vin[sampling_indices]

    return t_combined, Iout_noisy, Vin, Rsh_true, Rs_true, Ceq_true, Iph_true



# Define the neural network
class FCN(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN, N_LAYERS):
        super().__init__()
        activation = nn.Tanh
        self.fcs = nn.Sequential(
            nn.Linear(N_INPUT, N_HIDDEN),
            activation()
        )
        self.fch = nn.Sequential(
            *[nn.Sequential(
                nn.Linear(N_HIDDEN, N_HIDDEN),
                activation()
            ) for _ in range(N_LAYERS - 1)]
        )
        self.fce = nn.Linear(N_HIDDEN, N_OUTPUT)

    def forward(self, x):
        x = self.fcs(x)
        x = self.fch(x)
        x = self.fce(x)
        return x

def plot_errors(Rsh_true, Rs_true, Ceq_true, Iph_true, Rsh_est, Rs_est, Ceq_est, Iph_est):
    errors = {
        'Rsh': abs((Rsh_true - Rsh_est) / Rsh_true) * 100,
        'Rs': abs((Rs_true - Rs_est) / Rs_true) * 100,
        'Ceq': abs((Ceq_true - Ceq_est) / Ceq_true) * 100,
        'Iph': abs((Iph_true - Iph_est) / Iph_true) * 100
    }

    plt.figure(figsize=(8, 6))
    plt.bar(errors.keys(), errors.values(), color=['blue', 'green', 'red', 'orange'])
    plt.ylabel('Percentage Error (%)')
    plt.title('Percentage Error in Estimated Parameters')
    plt.show()


def plot_training_history(loss_history):
    epochs = np.arange(0, len(loss_history) * 500, 500)
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, loss_history, label="Training Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training Loss History")
    plt.grid(True)
    plt.legend()
    plt.show()

def plot_individual_parameter_history(history, true_value, label, color, scale_factor=1.0):
    """Plot individual parameter history with optional scaling."""
    epochs = np.arange(0, len(history) * 500, 500)
    scaled_history = [h * scale_factor for h in history]
    scaled_true_value = true_value * scale_factor
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, scaled_history, label=f'Estimated {label} (scaled)' if scale_factor != 1.0 else f'Estimated {label}', color=color)
    plt.axhline(y=scaled_true_value, color=color, linestyle='--', label=f'True {label} (scaled)' if scale_factor != 1.0 else f'True {label}')
    plt.xlabel("Epochs")
    plt.ylabel(f"{label} Value")
    plt.title(f"Evolution of {label} During Training")
    plt.xticks(fontsize=16)
    plt.legend(fontsize=20)  #  1 
    plt.yticks(fontsize=16)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot functions for results and errors
def plot_results(t_combined, Iout_noisy, Iout_pred):
    # Convert time scale to microseconds
    t_micro = t_combined.detach().numpy() * 1e-2

    plt.figure(figsize=(8, 6))
    plt.scatter(t_micro, Iout_noisy.detach().numpy(), label="Current Data", color="blue", s=10)
    plt.plot(t_micro, Iout_pred.detach().numpy(), label="Predicted Current", color='red')
    plt.xlabel('Time (s)')  # Change the label to microseconds
    plt.ylabel('Current (A)')
    plt.title('Comparison of Predicted Iout and Data')
    plt.legend()
    plt.grid(True)
    plt.show()

def train_PINN_with_scaled_Rsh(G, T, noise_level, epochs=250001, lr=5e-4, lambda1=1e4, lambda_l2=1e-8, hidden_layers=3, hidden_units=30):
    t_combined, Iout_noisy, Vin, Rsh_true, Rs_true, Ceq_true, Iph_true = generate_data(G, T, noise_level)
    pinn = FCN(1, 1, hidden_units, hidden_layers)

    Rsh_eq = torch.nn.Parameter(torch.ones(1, dtype=torch.float32, requires_grad=True)*0.86)
    Rs = torch.nn.Parameter(torch.tensor([1.6], requires_grad=True))
    Ceq = torch.nn.Parameter(torch.tensor([1.0], requires_grad=True))
    Iph = torch.nn.Parameter(torch.tensor([4.0], requires_grad=True))

    optimizer = torch.optim.AdamW([
        {'params': pinn.parameters()},
        {'params': [Rs, Ceq, Iph]},
        {'params': [Rsh_eq], 'lr': lr}
    ], lr=lr, weight_decay=lambda_l2)

    loss_history = []
    Rsh_history, Rs_history, Ceq_history, Iph_history = [], [], [], []
    lr_updated = False

    for epoch in range(epochs):
        optimizer.zero_grad()
        Vc = pinn(t_combined)
        Vc_dot = torch.autograd.grad(Vc, t_combined, torch.ones_like(Vc), create_graph=True)[0]

        loss1 = torch.mean((Vc_dot - ((Vin - Vc * (1 + Rs / (Rsh_eq * 100))) / (Rs * Ceq) + Iph / Ceq))**2)
        Iout_pred = (Vc - Vin) / Rs
        loss2 = torch.mean((Iout_pred - Iout_noisy)**2)
        loss = loss1 + lambda1 * loss2
        loss.backward()

        if not lr_updated and (Rsh_eq * 100).item() >= 92:
            for param_group in optimizer.param_groups:
                if id(Rsh_eq) in [id(p) for p in param_group['params']]:
                    param_group['lr'] = 5e-7
            print(f"Learning rate for Rsh_eq updated to 1e-7 at epoch {epoch} when Rsh_eq reached 95.")
            lr_updated = True

        optimizer.step()

        if epoch % 500 == 0:
            loss_history.append(loss.item())
            Rsh_history.append((Rsh_eq * 100).item())
            Rs_history.append(Rs.item())
            Ceq_history.append(Ceq.item())
            Iph_history.append(Iph.item())

    with torch.no_grad():
        V_pred = pinn(t_combined)
        Iout_pred = (V_pred - Vin) / Rs

    plot_results(t_combined, Iout_noisy, Iout_pred)
    plot_individual_parameter_history(Rsh_history, Rsh_true, "Rsh", "blue")
    plot_individual_parameter_history(Rs_history, Rs_true, "Rs", "green")
    plot_individual_parameter_history(Ceq_history, Ceq_true, "Ceq", "red", scale_factor=0.00001)  #  Ceq
    plot_individual_parameter_history(Iph_history, Iph_true, "Iph", "orange")


    # Display the final estimated value for Ceq scaled by 0.0001
    Ceq_est_scaled = Ceq.item() * 0.00001
    print("Final Estimated Ceq (scaled):", Ceq_est_scaled)

    return (Rsh_eq * 100).item(), Rs.item(), Ceq_est_scaled, Iph.item()

# Example usage
Rsh_est, Rs_est, Ceq_est, Iph_est = train_PINN_with_scaled_Rsh(G=1000, T=298, noise_level=0)
print("Estimated Rsh:", Rsh_est)
print("Estimated Rs:", Rs_est)
print("Estimated Ceq (scaled):", Ceq_est)
print("Estimated Iph:", Iph_est)
```
